### Parsowanie strony z opiniami

```python
page_dom = BeautifulSoup(response.text, 'html.parser')
print(type(page_dom))
```

```python
opinions = page_dom.select("div.js_product-review:not(.user-post--highlight)")
print(type(opinions))
print(len(opinions))
```

```python
opinion = page_dom.select_one("div.js_product-review")
print(type(opinion))
```

### Analiza struktury pojedynczej opinii

|składowa|nazwa|selektor|
|--------|-----|--------|
|opinia|opinion|div.js_product-review|
|identyfikator|opinion_id|[data-entry-id]|
|autor|author|span.user-post__author-name|
|treść|content|div.user-post__text|
|ocena|score|span.user-post__score-count|
|rekomendacja|recommendation|span.user-post__author-recomendation > em|
|lista zalet|pros|div.review-feature__item--positive|
|lista wad|cons|div.review-feature__item--negative|
|dla ilu przydatna|thumbs_up|button.vote-yes > span|
|dla ilu nieprzydatna|thumbs_down|button.vote-no > span|
|data zamieszczeenia|post_date|span.user-post__published > time:nth-child(1)[datetime]|
|data zakupu|purchase_date|span.user-post__published > time:nth-child(2)[datetime]|

```python
opinion = opinions.pop()
reviews = []    
review_data = {}
    
# Id opinii
review_data["id"] = opinion.get("data-entry-id")

# Autor
author_tag = opinion.select_one(".user-post__aisummary-author-name")
review_data["author"] = author_tag.text.strip() if author_tag else None

# Treść
content_tag = opinion.select_one(".user-post__text")
review_data["content"] = content_tag.text.strip() if content_tag else None

# Ocena
rating_tag = opinion.select_one(".user-post__score meta[itemprop='ratingValue']")
review_data["rating"] = float(rating_tag["content"]) if rating_tag else None

# Rekomendacja
rec_tag = opinion.select_one(".user-post__recommendation")
review_data["recommendation"] = rec_tag.text.strip() if rec_tag else None

# Zalety
pros = opinion.select(".review-feature__pros li")
review_data["pros"] = [p.text.strip() for p in pros] if pros else []

# Wady
cons = opinion.select(".review-feature__cons li")
review_data["cons"] = [c.text.strip() for c in cons] if cons else []

# Dla ilu przydatna
useful_tag = opinion.select_one(".vote-yes span[id^='votes-yes-']")
review_data["useful"] = int(useful_tag.text.strip()) if useful_tag else 0

# Dla ilu nieprzydatna
useless_tag = opinion.select_one(".vote-no span[id^='votes-no-']")
review_data["not_useful"] = int(useless_tag.text.strip()) if useless_tag else 0

# Data zamieszczenia
date_tag = opinion.select_one(".user-post__published .user-post__published-date")
review_data["published_date"] = date_tag.text.strip() if date_tag else None

# Data zakupu
purchase_tag = opinion.select_one(".user-post__purchased .user-post__published-date")
review_data["purchase_date"] = purchase_tag.text.strip() if purchase_tag else None

reviews.append(review_data)

# Wyświetlenie zebranych opinii
for r in reviews:
    print(r)
```
